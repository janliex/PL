```python
#dict字典
#用法：{key1:value1, key2:value2}

x =[["釋字第813號","文化資產保存法..."],["釋字第812號","中華民國94年2月2日修正..."],["釋字第811號","中華民國94年1月19日修正.."]]
print(dict(x))        

y = dict(a="文化資產保存法...", b="中華民國94年2月2日修正...", c="中華民國94年1月19日修正..")
print(y)
```

```python
x = {"釋字第813號": "文化資產保存法...",
     "釋字第812號": "中華民國94年2月2日修正...", 
     "釋字第811號": "中華民國94年1月19日修正.."}
print(x["釋字第813號"])
print(x.get("釋字第813號"))
print("----------------------------")
print(x.keys()) 
print(x.values())
print(x.items())
```

```python
#dict新增 
x = {"釋字第813號": "文化資產保存法..."}
new = {"釋字第812號": "中華民國94年2月2日修正...", 
       "釋字第811號": "中華民國94年1月19日修正.."}
x.update(new)#可用於兩個以上
print(x)
print("----------------------------")

#新增
x["釋字第810號"] = "不動產估價師法..."
print(x)
print("----------------------------")

#修改
x["釋字第810號"] = "原住民族工作權保障法..."
print(x)
```

```python
#刪除
x = {"釋字第813號": "文化資產保存法...",
     "釋字第812號": "中華民國94年2月2日修正...", 
     "釋字第811號": "中華民國94年1月19日修正.."}
del x["釋字第813號"]
print(x)

y = x.pop("釋字第812號")
print(y) #pop的value
print(x) #結果
```
