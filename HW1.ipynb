```python
#比較不同性別與最喜歡喝的飲料種類
import pandas as pd
data = pd.read_csv("Survey.csv")
print(data)
```

```python
#總數
gender = list(data["您的性別"])
print("總共人數:", len(data))
```

```python
#試印出單一類別的資訊
print(data["您的性別"])
```

```python
#試印出單一類別
girl = data[data['您的性別'].str.contains("生理女")]
girl.tail(4)
```

```python
tea = data[data['最喜歡喝的飲料種類'].str.contains("一般茶類")]
tea.tail(4)
```

```python
#交集(生理女 且 喜歡喝一般茶類)
data[(data["您的性別"]=="生理女") & (data["最喜歡喝的飲料種類"] == "一般茶類")]
```

```python
#聯集 (生理女 或 喜歡一般茶類)
data[(data["您的性別"]=="生理女") | (data["最喜歡喝的飲料種類"] == "一般茶類")]
```

```python
#差集(生理女 但不是 最喜歡一般茶類)
data[(data["您的性別"]=="生理女") != (data["最喜歡喝的飲料種類"] == "一般茶類")]
```
