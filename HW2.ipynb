```python
import json
```

```python
test = open('people .json',encoding="utf8")
Data = json.load(test)
type(Data)
```

```python
Data.items()
```

```python
temp = Data.get('台灣地區')
```

```python
ask = input("請輸入想查詢地區：")
```

```python
 for i in temp:
    if ask == i.get('site_id'):
        print("人口密度為:", i.get("population_density"))
```

```python
density = []
for i in temp:
    density.append(i.get("population_density"))
density
```

```python
density_enumerate = enumerate(density, start = 1)
```

```python
density_enumerate_sorted = sorted(density_enumerate, key = lambda s: s[1], reverse = True)
density_enumerate_sorted
```

```python
density_enumerate_sorted[0][0]
```

```python
for i in density_enumerate_sorted:
    j = i[0] - 1
    print(temp[j].get('site_id'), '人口密度是', temp[j].get('population_density'))
```
